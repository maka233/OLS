## 实现逻辑回归

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X = X[y<2,:2]
y = y[y<2]
```

```python
X.shape
```

```python
y.shape
```

```python
plt.scatter(X[y==0,0], X[y==0,1], color="red")
plt.scatter(X[y==1,0], X[y==1,1], color="blue")
plt.show()
```

### 使用逻辑回归

```python
from ML2021.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, seed=555)
```

```python
from ML2021.LogisticRegression import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
log_reg.predict_proba(X_test)
```

```python
log_reg.predict(X_test)
```
