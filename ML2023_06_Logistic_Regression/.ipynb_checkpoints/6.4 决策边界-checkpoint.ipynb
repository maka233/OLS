## 6.4 决策边界

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
X = X[y<2,:2]
y = y[y<2]
```

```python
plt.scatter(X[y==0,0], X[y==0,1], color="red")
plt.scatter(X[y==1,0], X[y==1,1], color="blue")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
from ML2021.LogisticRegression import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.coef_
```

## 线性决策边界

$$
x_{2} = \frac{-\theta_{0}-\theta_{1}x_{1}}{\theta_{2}}
$$

```python
def x2(x1):
    return (-log_reg.intercept_-log_reg.coef_[0]*x1)/log_reg.coef_[1]
```

```python
x1_plot = np.linspace(4, 8, 1000)
x2_plot = x2(x1_plot)
```

```python
plt.scatter(X_test[y_test==0,0], X_test[y_test==0,1], color="red")
plt.scatter(X_test[y_test==1,0], X_test[y_test==1,1], color="blue")
plt.plot(x1_plot,x2_plot)
plt.show()
```

### 非线性决策边界

```python
def plot_decision_boundary(model, axis):
    
    x0, x1 = np.meshgrid(
        np.linspace(axis[0], axis[1], int((axis[1]-axis[0])*100)).reshape(-1, 1),
        np.linspace(axis[2], axis[3], int((axis[3]-axis[2])*100)).reshape(-1, 1),
    )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_predict = model.predict(X_new)
    zz = y_predict.reshape(x0.shape)

    from matplotlib.colors import ListedColormap
    custom_cmap = ListedColormap(['#EF9A9A','#FFF59D','#90CAF9'])
    
    plt.contourf(x0, x1, zz, cmap=custom_cmap)
```

```python
plot_decision_boundary(log_reg, axis=[4, 7.5, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

### KNN的决策边界

对于kNN算法来说，k越小，决策边界越不规则，越过拟合。

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train,y_train)
```

```python
plot_decision_boundary(knn_clf, axis=[4, 7.5, 1.5, 4.5])
plt.scatter(X[y==0,0], X[y==0,1])
plt.scatter(X[y==1,0], X[y==1,1])
plt.show()
```

### 测试不同K值的影响

```python
knn_clf_all = KNeighborsClassifier(n_neighbors = 1)
knn_clf_all.fit(iris.data[:,:2],iris.target)
```

```python
plot_decision_boundary(knn_clf_all, axis=[4, 8, 1.5, 4.5])
plt.scatter(iris.data[iris.target==0,0], iris.data[iris.target==0,1])
plt.scatter(iris.data[iris.target==1,0], iris.data[iris.target==1,1])
plt.scatter(iris.data[iris.target==2,0], iris.data[iris.target==2,1])
plt.show()
```

```python
knn_clf_all_5 = KNeighborsClassifier(n_neighbors = 5)
knn_clf_all_5.fit(iris.data[:,:2],iris.target)
plot_decision_boundary(knn_clf_all_5, axis=[4, 8, 1.5, 4.5])
plt.scatter(iris.data[iris.target==0,0], iris.data[iris.target==0,1])
plt.scatter(iris.data[iris.target==1,0], iris.data[iris.target==1,1])
plt.scatter(iris.data[iris.target==2,0], iris.data[iris.target==2,1])
plt.show()
```

```python
knn_clf_all_25 = KNeighborsClassifier(n_neighbors = 25)
knn_clf_all_25.fit(iris.data[:,:2],iris.target)
plot_decision_boundary(knn_clf_all_25, axis=[4, 8, 1.5, 4.5])
plt.scatter(iris.data[iris.target==0,0], iris.data[iris.target==0,1])
plt.scatter(iris.data[iris.target==1,0], iris.data[iris.target==1,1])
plt.scatter(iris.data[iris.target==2,0], iris.data[iris.target==2,1])
plt.show()
```

```python
knn_clf_all_50 = KNeighborsClassifier(n_neighbors = 50)
knn_clf_all_50.fit(iris.data[:,:2],iris.target)
plot_decision_boundary(knn_clf_all_50, axis=[4, 8, 1.5, 4.5])
plt.scatter(iris.data[iris.target==0,0], iris.data[iris.target==0,1])
plt.scatter(iris.data[iris.target==1,0], iris.data[iris.target==1,1])
plt.scatter(iris.data[iris.target==2,0], iris.data[iris.target==2,1])
plt.show()
```
